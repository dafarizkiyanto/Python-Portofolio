```python
# Mengonversi kolom menjadi numerik, mengubah nilai yang tidak bisa dikonversi menjadi NaN
df['streams'] = pd.to_numeric(df['streams'], errors='coerce').fillna(0).astype(int)
df.info()
```

# **TOP 10 SONGS**

```python
top_10_songs = df.groupby('track_name')["streams"].sum()
top_10_songs = top_10_songs.sort_values(ascending=False).head(10).reset_index()
top_10_songs.columns = ['Song Title', 'Total Streams']
top_10_songs
```

```python
colors = plt.cm.viridis(np.linspace(0, 1, len(top_10_songs)))

plt.figure(figsize=(10,6))
plt.barh(top_10_songs['Song Title'], top_10_songs['Total Streams'],color=colors)
plt.xlabel('Total Streams')
plt.ylabel('Song Title')
plt.title('Top 10 Songs by Total Streams')
plt.gca().invert_yaxis()
plt.show()
```

# **Top 10 Artist with Most Stream**

```python
top_10_artist = df.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).head(10)
top_10_artist = top_10_artist.reset_index()
top_10_artist.columns = ['Artist Name', 'Total Streams']
top_10_artist
```

```python
plt.figure(figsize=(10, 6))
plt.barh(top_10_artist['Artist Name'], top_10_artist['Total Streams'], color='skyblue')
plt.xlabel('Total Streams')
plt.ylabel('Artist Name')
plt.title('Top 10 Artists by Total Streams')
plt.gca().invert_yaxis()
```

# **Total Songs Released by Year**

```python
song_by_year = df.groupby(df['released_year']).size().reset_index(name='Total Songs')
song_by_year = song_by_year.sort_values(by='Total Songs',ascending=False).head(10).reset_index(drop=True)
song_by_year.columns = ['Year', 'Total Songs']
song_by_year
```

```python
plt.figure(figsize=(10, 6))
plt.barh(song_by_year['Year'], song_by_year['Total Songs'], color='lightcoral')
plt.xlabel('Total Songs')
plt.ylabel('Year')
plt.title('Total Songs Released by Year')
```

# **TOP 10 MOST STREAMED SONG (FEATURED ARTISTS)**

```python
top_10_featured_songs = (
    df[df['artist_count'] >= 2]
    .groupby(['track_name', 'artist(s)_name'])['streams']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

# Mengatur nama kolom
top_10_featured_songs.columns = ['Song Title', 'Artist(s)', 'Total Streams']
top_10_featured_songs
```

```python
# Gabungkan kolom 'Song Title' dan 'Artist(s)' menjadi satu kolom baru untuk label y
top_10_featured_songs['Song & Artist'] = top_10_featured_songs['Song Title'] + " - " + top_10_featured_songs['Artist(s)']

# Plot
plt.figure(figsize=(10, 6))
plt.barh(top_10_featured_songs['Song & Artist'], top_10_featured_songs['Total Streams'], color='lightgreen')
plt.xlabel('Total Streams')
plt.title('Top 10 Most Streamed Songs (Featured Artists)')
plt.gca().invert_yaxis()
plt.show()
```

# **Artist Energy Distribution**

```python

def plot_artist_energy_distribution(df, artist_name):
    # Filter songs for the specified artist
    artist_songs = df[df['artist(s)_name'] == artist_name]

    # Check if the artist has any songs in the dataset
    if artist_songs.empty:
        print(f"No songs found for artist: {artist_name}")
        return

    # Create the histogram plot
    plt.figure(figsize=(8, 6))
    sns.histplot(artist_songs['energy_%'], kde=True, color='green')
    plt.title(f'Distribution of Song Energy for {artist_name}')
    plt.xlabel('Energy (%)')
    plt.ylabel('Count')
    plt.show()

# Example usage
plot_artist_energy_distribution(df, "Billie Eilish")
```

## **ARTIST STREAM MOST OVER TIME**

```python
df['release_date'] = pd.to_datetime(df['released_year'].astype(str) + '-' +
                                    df['released_month'].astype(str) + '-' +
                                    df['released_day'].astype(str))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_artist_streams_over_time(df, artist_name):
    # Filter the DataFrame for the specified artist
    artist_df = df[df['artist(s)_name'] == artist_name]

    # Check if the artist has any songs in the dataset
    if artist_df.empty:
        print(f"No songs found for artist: {artist_name}")
        return

    # Group by release date (monthly) and aggregate total streams
    artist_streams = artist_df.groupby(artist_df['release_date'].dt.to_period('M')).agg({'streams': 'sum'}).reset_index()
    artist_streams['release_date'] = artist_streams['release_date'].dt.to_timestamp()

    # Create the line plot
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='release_date', y='streams', data=artist_streams)
    plt.title(f'Total Streams Over Time for {artist_name}')
    plt.xlabel('Release Date')
    plt.ylabel('Streams')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout for better spacing
    plt.show()

# Example usage
plot_artist_streams_over_time(df, 'The Weeknd')
```
